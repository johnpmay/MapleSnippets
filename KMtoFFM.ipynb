# Verify GF2 Matrix Multiplication Formulas of Kauers and Moosbauer

This Maple notebook downloads, converts, and verifies the matrix multiplication formulas reported in https://arxiv.org/pdf/2210.04045.pdf

```maple
with(StringTools): with(LinearAlgebra): interface(rtablesize=16):
```

This is the file for the 4x4 product using 47 multiplications

```maple
filestring := Import(
    "http://www.algebra.uni-linz.ac.at/people/mkauers/matrix-mult/s47.exp", 
    format="text"
   ):
n := 4;   
```

```maple
prods := map(parse, Split(Trim(filestring),"\n") );
```

```maple
prods[1];
```

The following procedures parse the products in the .exp files into the format used by the FMM catalog at https://fmm.univ-lille.fr/

```maple
prodtotriad := proc(expr::`*`, n)
local a,b,c;
    if nops(expr)<>3 then
        error;
     end if;
    (a,b,c) := op(expr);
    return [sumtomatrix(a,n), sumtomatrix(b,n), sumtomatrix(c,n)];
end proc:

sumtomatrix := proc(expr::{`+`,name}, n)
local am, t, i, j;
    am := Matrix(n,n):
    for t in [op(expr)] do
        i := sscanf(substring(t,2),"%d");
        j := sscanf(substring(t,3),"%d");
        am[i,j] := 1;
    end do:
    return am;
end proc:
```

```maple
Tensor := TriadSet([seq(Triad(prodtotriad(p, n)),p in prods)]):
```

```maple
V:=Vector(op(Tensor)):
V[1..10],V[11..20],V[21..30],V[31..40],V[41..47];
```

```maple
dims := [n,n,n];
```

```maple
rank := nops(prods);
```

```maple
A := Matrix(dims[1], dims[2], (i,j)->cat(`a__`,i,`,`,j)):
B := Matrix(dims[2], dims[3], (i,j)->cat(`b__`,i,`,`,j)):
C := Matrix(dims[1], dims[3], (i,j)->cat(`c__`,i,`,`,j)):
A,B,C;
```

```maple
chk := add(
        Trace(Transpose(op([1,i,1,1],Tensor)).A)
        * Trace(Transpose(op([1,i,1,2],Tensor)).B)
        * Transpose(op([1,i,1,3],Tensor)),
    i=1..rank):
```

```maple
(map(expand,A.B - chk) mod 2)
```

```maple
tmp := seq(
    LinearAlgebra:-Trace(LinearAlgebra:-Transpose(op([1,i,1,1],Tensor)).A)
    *LinearAlgebra:-Trace(LinearAlgebra:-Transpose(op([1,i,1,2],Tensor)).B)
    *LinearAlgebra:-Transpose(op([1,i,1,3],Tensor)),
    i=1..rank):
```

```maple
MUL := [seq(
        cat(`m__`,i)=remove(type,convert(tmp[i],set)[-1],numeric),
    i=1..rank)]:
Vector(MUL);
```

```maple
ADD := [seq(seq(
    cat(`c__`,i,`,`,j)=
        add(
            ifelse(tmp[k][i,j]=0, 0,
                select(type,tmp[k][i,j],numeric)*cat(`m__`,k)),
        k=1..rank),
        j=1..dims[3]),i=1..dims[1])]:
Vector(ADD);
```

```maple
`*`=nops(MUL), `+`=`+`(seq(nops(rhs(e))-1, e in ADD)) + `+`(seq(nops(op(1,rhs(e)))+nops(op(2,rhs(e)))-2, e in MUL));
```

The individual formulas for the entries of C can be relatively large, but expanded out and reduced modulo 2

```maple
(subs(subs(MUL,ADD),C[1,1]) mod 2);
```

Now do all the entries of C.  First compute the symbolic 4x4x4 matrix product A.B, then substitute the multiplication formulas in `MUL` into the addition formulas in `ADD`, and then substitute those into the Symbolic Matrix C, and check that A.B=C mod 2 after expanding

```maple
map(expand,A.B-subs(subs(MUL,ADD),C)) mod 2;
```
